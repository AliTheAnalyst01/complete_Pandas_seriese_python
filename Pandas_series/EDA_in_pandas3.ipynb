```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('world_population.csv')

df
```

```python
# set the number display in float at two decimal

pd.set_option('display.float_format',lambda x:'%.2f' % x)
df
```

```python
# now check the whole info about the dataframe
df.info()

```

```python
# here i get the high level overview of numerical data
df.describe()
```

```python
# here i check the null value in dataset
df.isnull().sum()
```

```python
# now saw the only null values portion of dataset
df[df['2022 Population'].isnull()]
```

```python
df_lowest_popu = df.sort_values(by='2022 Population').head(10).set_index('Rank')
df_lowest_popu
```

```python
highest_popu=df.sort_values(by='2022 Population',ascending=False).head(10).set_index('Rank')
highest_popu
```

```python
# check the corr of numerical value
df.corr()
```

```python
sns.heatmap(df.corr(),annot=True)
plt.rcParams['figure.figsize']=(25,10)
plt.show()
```

```python
# now i make the group on the basis on continent
df.groupby('Continent').mean()
```

```python
# now i sort the value on the basis of diferent columns
df2=df.groupby('Continent')['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population'].mean().sort_values(by='2022 Population',ascending=False)
df2
```
