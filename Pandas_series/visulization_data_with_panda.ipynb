```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Ice Cream Ratings.csv')
df
```

```python
# now we make the visulization by default
df.plot()
```

#### as we saw the by default pandas make the line plot

```python
# when i want to saw the subplot of the any graph or this graph
df.plot(kind='line',subplots=True)
```

##### as we saw that subplot of line graph individually

```python
# i want to give the title to the graph
df.plot(kind='line',title='Ice Cream Ratings')
```

```python
# mention the axis
df.plot(kind='line',x='Date',y=['Flavor Rating','Texture Rating','Overall Rating'])
```

```python
df.plot(kind='line',x='Date',y=['Flavor Rating','Texture Rating','Overall Rating'],
        title="Ice Cream Ratings.csv",
        xlabel='Daily Ratings',
        ylabel='Score')
```

```python
# now for the bar chart 
df.plot(kind='bar')
```

```python
df.plot(kind='bar',stacked=True)
```

```python
df['Flavor Rating'].plot(kind='bar',stacked=True,x='Date',y='Flavor Rating')
```

```python
# for horizantal bar chart 
df.plot.barh(stacked=True)
```

```python
# now for the scatter plot
# we will use the same data as above
df.plot.scatter(x='Flavor Rating',y='Texture Rating')
```

```python
# now going for sizw of dot 
df.plot.scatter(x='Flavor Rating',y='Texture Rating',s=500)
```

```python
# now going to check the color of graph point
df.plot.scatter(x='Flavor Rating',y='Texture Rating',s=500,c='black')
```

```python
# now  we make the histogram 
df.plot.hist()
```

```python
# now we make the boxplot
df.boxplot()
```

```python
# area chart 
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y='Flavor Rating')
```

```python
df.plot.pie(y='Flavor Rating',figsize=(10,10))
```
